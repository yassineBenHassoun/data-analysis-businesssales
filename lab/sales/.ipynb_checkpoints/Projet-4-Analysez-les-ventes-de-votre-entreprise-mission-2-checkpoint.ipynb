```python
# import des librairies dont nous aurons besoin
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
import scipy.stats as st
from sklearn import linear_model
import scipy.stats
from scipy.stats import chi2_contingency
from scipy.stats import iqr
# chargement et affichage des données
ventes = pd.read_csv("C:/Users/cleme/OneDrive/Bureau/Openclassroom/Projet 4/Cours/ventes.csv")
print(ventes)
```

```python
# Mission 2 : Analyse des données 
# Nous allons dans une premier temps analyser nos ventes, puis analyser nos clients pour mieux les connaitres et également connaitre nos livres vendus
```

```python
ventes['date']= pd.to_datetime(ventes['date'])
ventes['date2']= ventes['date'].dt.strftime('%d-%m-%Y') 
ventes['date3']= ventes['date'].dt.strftime('%Y-%m')
```

```python
ventes.head()
```

```python
# Dans la mission 1 en calculant les âges nous nous sommes rendu comptes que les ventes de notre dataframe sont de mars 2021 à février 2022
```

```python
chiffredaffaire = ventes[["date3", "price"]].groupby("date3", as_index=False).sum().sort_values("date3",ascending=True)
ax = chiffredaffaire.plot(kind = "bar", figsize=(12,5),color="green")
ax.set_xlabel("date", fontsize=16)
ax.set_ylabel("ca", fontsize=16)
ax.xaxis.set_ticklabels(['mars', 'avril', 'mai', 'juin', 'juillet','août','septembre','octobre','novembre','décembre','janvier','février'], rotation = 0, color = 'black', fontsize = 10, style = 'normal', verticalalignment = 'center')
ax.set_title("chiffre d'affaires par mois", fontsize=16)
```

```python
# Le CA est assez stable à part le mois d'octobre où les ventes ont baissé.
# Nous allons donc analyser plus en details
```

```python
CAparcategorie = ventes[ventes['categ'] != '0']

graph = ventes.groupby(['categ', 'date3'])['price'].sum().unstack('categ').fillna(0)
ax = graph.plot(kind = "bar", stacked=True, figsize=(10,10))
ax.set_xlabel("date", fontsize=16)
ax.set_ylabel("ca", fontsize=16)
ax.xaxis.set_ticklabels(['mars', 'avril', 'mai', 'juin', 'juillet','août','septembre','octobre','novembre','décembre','janvier','février'], rotation = 90, color = 'black', fontsize = 12, style = 'normal', verticalalignment = 'top')
ax.set_title("chiffre d'affaires par mois / catégories cumulées", fontsize=16)
plt.savefig('CA par mois catégories cumulées.png')
```

```python
# nous allons maintenant analyser grâce à la courbe de Lorenz si les acheteurs dépenses tous de la même manière?

depenses = ventes[ventes['price']>0]
dep = depenses['price'].values
lorenz = np.cumsum(np.sort(dep)) / dep.sum()
lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0
plt.plot([0,1], [0,1], color='blue') #tracé de la bissectrice
plt.plot(np.linspace(0,1,len(lorenz)),lorenz,drawstyle='steps-post',color='green') #tracé courbe de Lorenz
plt.title('courbe de Lorenz des dépenses des acheteurs')
plt.savefig('courbe de lorenz pour les acheteurs.png')
plt.show()
```

```python
# Aire sous la courbe de Lorenz. La dernière valeur ne participe pas à l'aire, d'où : "[:-1]"
aire_ss_courbe = lorenz[:-1].sum()/len(lorenz) 

# Aire entre la 1e bissectrice et la courbe de Lorenz.
S = 0.5 - aire_ss_courbe 
gini = 2*S

gini
```
