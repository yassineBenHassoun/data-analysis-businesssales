```python
data 
s =  data.groupby(['sex','month']).sum().reset_index()
s

monthName = ['janv-2K22','fev-2K22','mars','avril','mai','juin','juillet','aout','septembre','octobre','novembre','decembre']

f = s.loc[sexMonth['sex'] == 'f']
h = s.loc[sexMonth['sex'] == 'm']
h['m_s'] = monthName
f['m_s'] = monthName
```
