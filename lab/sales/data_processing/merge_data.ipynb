```python
import pandas as pd
import numpy as np
from os import getcwd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import calendar
```

```python
transactions = pd.read_csv("../data_values_clear/transactions.csv")
clients = pd.read_csv("../data_values_clear/clients.csv")
products = pd.read_csv("../data_values_clear/products.csv")
```

```python
customersTransaction = clients.merge(transactions, on="client_id")
```

```python
mergeAll = customersTransaction.merge(products,on="id_product")
mergeAll['user_age'] = mergeAll['sell_year'] -  mergeAll['user_birthday']
```

```python
##binning age
mergeAll['user_age'] = mergeAll['user_age'].astype(int)
mergeAll['binning_age'] = pd.cut(x=mergeAll['user_age'], bins=[18,20,30,40,50,60,70,80,90,100])
mergeAll.dtypes
```

```python
mergeAll.loc[mergeAll['month'] == 1,'month_string'] = 'Janvier - 2022'
mergeAll.loc[mergeAll['month'] == 2,'month_string'] = 'Fevrier - 2022'
mergeAll.loc[mergeAll['month'] == 3,'month_string'] = 'Mars'
mergeAll.loc[mergeAll['month'] == 4,'month_string'] = 'Avril'
mergeAll.loc[mergeAll['month'] == 5,'month_string'] = 'Mai'
mergeAll.loc[mergeAll['month'] == 6,'month_string'] = 'Juin'
mergeAll.loc[mergeAll['month'] == 7,'month_string'] = 'Juillet'
mergeAll.loc[mergeAll['month'] == 8,'month_string'] = 'Aout'
mergeAll.loc[mergeAll['month'] == 9,'month_string'] = 'Septembre'
mergeAll.loc[mergeAll['month'] == 10,'month_string'] = 'Octobre'
mergeAll.loc[mergeAll['month'] == 11,'month_string'] = 'Novembre'
mergeAll.loc[mergeAll['month'] == 12,'month_string'] = 'Decembre'

mergeAll
```

```python
mergeAll['traceability_transaction'] = 1
```

```python
mergeAll
```

```python
mergeAll = mergeAll[['client_id', 'sex','user_birthday','binning_age','user_age','category_id','id_product'
                           ,'price','sell_date','transaction_date','sell_year','month','month_string','traceability_transaction'
                          ]];

resultMerge = mergeAll.reset_index()
resultMerge
```

```python
resultMerge.to_csv("../data_values_clear/merge.csv", index = False, header=True)
```
