```python
import pandas as pd
import numpy as np
from os import getcwd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
import calendar
```

```python
datas = pd.read_csv("../sales/data-clear/all-data.csv",index_col="transaction_date", parse_dates=True)
datas['year_transaction'] = pd.DatetimeIndex(datas['sell_date']).year
datas['y_m_transaction'] = pd.to_datetime(datas['sell_date']).dt.strftime('%Y-%m')
datas['m_transaction'] = pd.to_datetime(datas['sell_date']).dt.strftime('%m')
data['user_age'] = data['user_age'].astype(int)
data['binning_age'] = pd.cut(x=data['user_age'], bins=[18,20,30,40,50,60,70,80,90,100])
data['month_t'] = pd.to_datetime(data['sell_date'], errors='coerce').dt.month
data.loc[data['month_t'] == 1,'month_transaction'] = 'Janvier'
data.loc[data['month_t'] == 2,'month_transaction'] = 'Fevrier'
data.loc[data['month_t'] == 3,'month_transaction'] = 'Mars'
data.loc[data['month_t'] == 4,'month_transaction'] = 'Avril'
data.loc[data['month_t'] == 5,'month_transaction'] = 'Mai'
data.loc[data['month_t'] == 6,'month_transaction'] = 'Juin'
data.loc[data['month_t'] == 7,'month_transaction'] = 'Juillet'
data.loc[data['month_t'] == 8,'month_transaction'] = 'Aout'
data.loc[data['month_t'] == 9,'month_transaction'] = 'Septembre'
data.loc[data['month_t'] == 10,'month_transaction'] = 'Octobre'
data.loc[data['month_t'] == 11,'month_transaction'] = 'Novembre'
data.loc[data['month_t'] == 12,'month_transaction'] = 'Decembre'
data['order'] = 1

data
```
